```python
from pyspark.sql import SparkSession
```

```python
spark = SparkSession.builder.appName('basic').getOrCreate()
```

```python
df = spark.read.csv("/home/sai/ex/DF/st.csv", inferSchema=True, header=True)
```

```python
df.printSchema()
```

```python
df.show()
```

```python
df.filter("Close < 500").select('Open').show()
```

```python
df.filter(df['Close']<500).select('Open').show()
```

```python
#Filtering based on multiple conditions. For NOT '~'
#for equal '==' | 'AND' = '&'
df.filter((df['Close'] < 200) & (df['Open'] > 100)).show()
```

```python
#TO save result of condition in a row object
result = df.filter(df['Low'] == 197.16).collect()
```
